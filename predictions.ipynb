```python
import pandas as pd
```

```python
# Load the races data from the CSV file into a DataFrame
races = pd.read_csv('races.csv', index_col=0)
```

```python
# Get the shape of the DataFrame (number of rows and columns)
races.shape
```

```python
# Count the occurrences of each driver in the 'Driver' column
races['Driver'].value_counts()
```

```python
races['Result_code_x'] = races['Result_x'].astype('category').cat.codes
races['Result_code_y'] = races['Result_y'].astype('category').cat.codes
races['Grid_code_x'] = races['Grid_x'].astype('category').cat.codes
races['Grid_code_y'] = races['Grid_y'].astype('category').cat.codes
races['Track_code'] = races['Race'].astype('category').cat.codes
```
